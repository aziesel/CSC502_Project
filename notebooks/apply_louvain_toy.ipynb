Rather, I then pursued the modularity calculation as laid out in a number of other papers, including the [Wikipedia article](https://en.wikipedia.org/wiki/Louvain_method) describing the Louvain algorithm. 

```python
#m = the sum of all weighted edges in the graph. Only need to collect once.
def get_m(graph):
  m = 0
  all_weights = aggregated.rdd.map(lambda x: x[3]).collect()
  all_weights = [item for sublist in all_weights for item in sublist]
  [m := m + x for x in all_weights]

  return m

#sigma_in, the sum of edge weights WITHIN cluster 
def get_sigma_in(cluster):
  volume = 0
  cluster_rows = aggregated.filter(aggregated.clusterID == cluster)
  for i in range(cluster_rows.count()):
    neighbours_set = cluster_rows.rdd.map(lambda x: str(x[2])).collect()
    neighbours_set = neighbours_set[0].lstrip("[").rstrip("]").split(",")
    neighbours_set = [x.replace("'", "").strip(" ") for x in neighbours_set]
    weights_set = cluster_rows.rdd.map(lambda x: x[3]).collect()
    weights_set = str(weights_set[0]).split(",")
    weights_set = [x.replace("[", "").replace("]", "") for x in weights_set]
    for j, neighbour in enumerate(neighbours_set):
      if get_cluster(neighbour) == cluster:
        volume += int(weights_set[j])

  return volume

#sigma_tot, the sum of all edge weights, including those that cross out of the cluster
def get_sigma_tot(cluster):
  volume = 0
  cluster_rows = aggregated.filter(aggregated.clusterID == cluster)
  for i in range(cluster_rows.count()):
    neighbours_set = cluster_rows.rdd.map(lambda x: str(x[2])).collect()
    neighbours_set = neighbours_set[0].lstrip("[").rstrip("]").split(",")
    neighbours_set = [x.replace("'", "").strip(" ") for x in neighbours_set]
    weights_set = cluster_rows.rdd.map(lambda x: x[3]).collect()
    weights_set = str(weights_set[0]).split(",")
    weights_set = [x.replace("[", "").replace("]", "") for x in weights_set]
    [volume := volume + int(x) for x in weights_set]

  return volume

def get_degree(node):
  degree = 0
  node_row = aggregated.filter(aggregated.node == node)  
  weights_set = node_row.rdd.map(lambda x: x[3]).collect()
  weights_set = str(weights_set[0]).split(",")
  weights_set = [x.replace("[", "").replace("]", "") for x in weights_set]
  [degree := degree + int(x) for x in weights_set]

  return degree

def get_dest_degree(node, cluster):
  dest_degree = 0
  node_row = aggregated.filter(aggregated.node == node)  
  neighbours_set = node_row.rdd.map(lambda x: str(x[2])).collect()
  neighbours_set = neighbours_set[0].lstrip("[").rstrip("]").split(",")
  neighbours_set = [x.replace("'", "").strip(" ") for x in neighbours_set]
  weights_set = node_row.rdd.map(lambda x: x[3]).collect()
  weights_set = str(weights_set[0]).split(",")
  weights_set = [x.replace("[", "").replace("]", "") for x in weights_set]
  for i, neighbour in enumerate(neighbours_set):
    if get_cluster(neighbour) == cluster:
      dest_degree += int(weights_set[i])

  return dest_degree 

def get_neighbours(node):
  node_row = aggregated.filter(aggregated.node == node)  
  neighbours_set = node_row.rdd.map(lambda x: str(x[2])).collect()
  neighbours_set = neighbours_set[0].lstrip("[").rstrip("]").split(",")
  neighbours_set = [x.replace("'", "").strip(" ") for x in neighbours_set]

  return neighbours_set

def get_cluster(node):
  node_row = aggregated.filter(aggregated.node == node) 
  node_cluster = node_row.rdd.map(lambda x: int(x[0])).collect()

  return node_cluster[0]

def calc_modularity(cluster, m):
  Q = 0
  first_term = (get_sigma_in(cluster)) / (2*m)
  second_term = ((get_sigma_tot(cluster)) / (2*m) ) **2
  Q = first_term - second_term

  return Q

def calc_delta_Q(node, dest_cluster, m):
  delta_Q = 0
  sigma_in = get_sigma_in(dest_cluster)
  sigma_tot = get_sigma_tot(dest_cluster)
  k_i = get_degree(node)
  k_i_in = get_dest_degree(node, dest_cluster)

  first_term = (sigma_in + 2*k_i_in)/(2*m)
  second_term = ((sigma_tot + k_i)/(2*m)) **2
  third_term = (sigma_in)/(2*m)
  fourth_term = ((sigma_tot)/(2*m)) **2
  fifth_term = ((k_i)/(2*m)) **2

  delta_Q = (first_term - second_term) - (third_term - fourth_term - fifth_term)

  return delta_Q
```

I then developed additional functions to run through the neighbour for a given node, identify the cluster membership for those neighbours, and assess the ΔQ score for relocating that particular node to each of its neighbouring clusters. To avoid doing needless work, the function `check_neighbour_clusters` keeps track of which clusters it has already calclulated ΔQ for. 

```python
#for a given node, calculate the delta_Q for adding it to each of its neighbours' clusters and return best destination

def check_neighbour_clusters(node):
  #aggregated.cache()
  already_checked = [] #don't recheck multinode clusters
  max_delta_Q = 0
  best_dest_cluster = -1
  for neighbour in get_neighbours(node):
    dest_cluster = get_cluster(neighbour)
    if dest_cluster not in already_checked:
      delta_Q = calc_delta_Q(node, dest_cluster, m)
      if delta_Q > max_delta_Q:
        max_delta_Q = delta_Q
        best_dest_cluster = dest_cluster
    already_checked.append(dest_cluster)
  if best_dest_cluster == -1:
    best_dest_cluster = get_cluster(node) #stay at current cluster
  print("for ", node, " best dest cluster ", best_dest_cluster, "delta Q ", max_delta_Q)

  return best_dest_cluster

#have to update aggregated df after each node-cluster reassignment
#structure: clusterID, node, neighbour, weights
def update_aggregated(node, new_cluster):
  updated_aggregated = aggregated.withColumn("clusterID", when(col("node") == node, new_cluster).otherwise(col("clusterID")))

  return updated_aggregated

def shuffle_aggregated():
  shuffle_aggregated = aggregated.sample(frac=1)

  return shuffle_aggregated
```

```python
#manually executed round 3
node_list = input_C.rdd.map(lambda x: x[0]).collect()

updated_clusters = []
for node in node_list:
  print(node, get_cluster(node))
  dest_cluster = check_neighbour_clusters(node)
  aggregated = update_aggregated(node, dest_cluster)
  updated_clusters.append(dest_cluster)
```
